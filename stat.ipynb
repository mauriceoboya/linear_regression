```python
import pandas as pd
import statsmodels.api as sms
import numpy as np
```

```python
internet = pd.read_csv('internetusage.csv')
internet.head(5)
```

```python
new_var = sms.add_constant(internet['bachelors_degree'])
new_var.head()
```

```python
model=sms.OLS(internet['internet_usage'],sms.add_constant(internet['bachelors_degree'])).fit()
model.summary()
```

```python
bach=float(int(10))
```

```python
prediction=model.predict(pd.DataFrame({'intercept': 1,'bach':[bach]}))
```
